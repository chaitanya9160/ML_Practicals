```python
import numpy as np
import matplotlib.pyplot as plt

# Create a simple dataset: a sine curve with some noise
np.random.seed(0)
X = np.linspace(-3, 3, 20)
y = np.sin(X) + np.random.normal(0, 0.3, X.shape)

# Dense test data for plotting true function and predictions
X_test = np.linspace(-3, 3, 300)
y_true = np.sin(X_test)

# Fit models of increasing complexity
models = {
    "Underfitting (Linear)": np.poly1d(np.polyfit(X, y, 1)),     # Degree 1
    "Good Fit (Cubic)": np.poly1d(np.polyfit(X, y, 3)),          # Degree 3
    "Overfitting (Degree 15)": np.poly1d(np.polyfit(X, y, 15)),  # Degree 15
}

# Plotting
plt.figure(figsize=(14, 6))

for i, (title, model) in enumerate(models.items(), 1):
    plt.subplot(1, 3, i)
    plt.scatter(X, y, color='blue', label='Training Data')
    plt.plot(X_test, y_true, color='green', label='True Function')
    plt.plot(X_test, model(X_test), color='red', linestyle='--', label='Model Prediction')
    plt.title(title)
    plt.xlabel('X')
    plt.ylabel('y')
    plt.legend()

plt.tight_layout()
plt.suptitle("Underfitting vs Good Fit vs Overfitting", fontsize=16, y=1.05)
plt.show()
```
